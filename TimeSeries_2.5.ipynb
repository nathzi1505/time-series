```python
import pandas as pd
```

```python
PATH = "../Data/starbucks.csv"
```

```python
df = pd.read_csv(PATH, index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df['Close'].plot()
```

```python
df['Volume'].plot()
```

```python
title = 'Close'
ylabel = 'Price'
xlabel = 'Date'
```

```python
ax = df['Close'].plot(figsize=(12,6), title=title)
ax.autoscale(axis='y', tight=True)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
df['Close']['2017-01-01':'2017-12-31'].plot(figsize=(12, 6))
```

```python
df['Close'].plot(figsize=(12,6), xlim=['2017-01-01','2017-12-31'])
```

```python
df['Close'].plot(figsize=(12,6), ylim=[40, 70], ls='--', c='r')
```

```python
from matplotlib import dates
```

```python
ax = df['Close'].plot(xlim=['2017-01-01','2017-03-01'], ylim=[50,60], figsize=(12,6))
ax.set(xlabel='')

ax.xaxis.set_major_locator(dates.WeekdayLocator(byweekday=0))
ax.xaxis.set_major_formatter(dates.DateFormatter("%d"))

ax.xaxis.set_minor_locator(dates.MonthLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter("\n\n%b"))

ax.yaxis.grid(True)
ax.xaxis.grid(True)
```
