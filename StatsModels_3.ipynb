```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "../Data/airline_passengers.csv"
```

```python
airline = pd.read_csv(PATH, index_col="Month")
```

```python
airline.dropna(inplace=True)
```

```python
airline.index = pd.to_datetime(airline.index)
```

```python
airline.head()
```

```python
# Simple Moving Average
airline['6-month-SWA'] = airline['Thousands of Passengers'].rolling(window=6).mean()
```

```python
# Simple Moving Average
airline['12-month-SWA'] = airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.plot(figsize=(12,6))
```

```python
# Exponential Moving Average
airline['EWMA-12'] = airline['Thousands of Passengers'].ewm(span=12).mean()
```

```python
airline[['Thousands of Passengers', 'EWMA-12']].plot(figsize=(12,6))
```
