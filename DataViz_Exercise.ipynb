```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = "../03-Pandas-Visualization/df3.csv"
```

```python
df = pd.read_csv(PATH)
```

```python
df.head()
```

```python
df.plot.scatter(x='produced', y='defective', c='r', s=25, figsize=(12, 3))
```

```python
df['produced'].plot.hist()
```

```python
df['produced'].plot.hist(edgecolor='k').autoscale(enable=True, axis='x', tight=True)
```

```python
df.boxplot(by='weekday', column='produced', figsize=(12, 10))
```

```python
df['defective'].plot.kde()
```

```python
df['defective'].plot.kde(lw=5, ls='--')
```

```python
df.loc[:30].plot.area(stacked=False, alpha=0.4)
```

```python
ax = df.loc[:30].plot.area(stacked=False, alpha=0.4)
ax.legend(loc=0, bbox_to_anchor=(1.3, 0.6))
```
