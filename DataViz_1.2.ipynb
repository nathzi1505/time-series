```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH_df2 = "../03-Pandas-Visualization/df2.csv"
```

```python
df2 = pd.read_csv(PATH_df2)
```

```python
df2.head()
```

```python
df2['c'].plot.line(figsize=(10, 3), ls=':', c='r', lw=2)
```

```python
title = "X vs Y"
xlabel = "X"
ylabel = "Y"

ax = df2['c'].plot.line(figsize=(10, 3), ls=':', c='r', lw=2, title=title)
ax.set(xlabel=xlabel, ylabel=ylabel)
```

```python
ax = df2.plot()
ax.legend(loc='lower right')
```
